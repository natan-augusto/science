## Theory
This program applies the theory of converting physical quantities (length, time, and mass) between different units of the International System of Units (SI).
According to the textbook:

>"Incorporar um prefixo a uma unidade do SI tem o efeito de multiplicar a unidade pelo fator correspondente."
>(Incorporating a prefix into an SI unit has the effect of multiplying the unit by the corresponding factor.)

For example:

- 1 kilometer (km) is $10^3$ meters, so to convert 5 km to meters, we calculate $5 \times 10^3 = 5000$ m.
- 1 milligram (mg) is $10^{-6}$ kilograms, so to convert 2500 mg to kg, we calculate $2500 \times 10^{-6} = 0.0025$ kg.

| $10^6$ | mega | M |
| $10^3$ | quilo | k |

| $10^1$ | deca | da |

- Submultiples ($Factor < 1$):

| Factor | Prefix | Symbol |
| :--- | :---: | ---: |
| $10^{-1}$ | deci | d |
| $10^{-2}$ | centi | c |
| $10^{-3}$ | mili | m |

This program automates this logic by mapping the exponent associated with a given prefix and performing the corresponding calculations.
