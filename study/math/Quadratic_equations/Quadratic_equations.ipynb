# Objective
This program is a mathematical application for quadratic equations; it functions as a root calculator for any quadratic equation with real roots.

---
# Theory

The program is a direct implementation of the standard method for solving a second-degree polynomial equation, also known as a quadratic equation, in its general form:

$$ax^2 + bx + c = 0$$

## The Discriminant (Δ)

It is one of the constants needed to find the roots of a quadratic equation and is calculated using the following formula:

$$\Delta = b^2 - 4ac$$

As can be inferred from the code structure, the discriminant directly affects the roots of the equation, which may be real or complex: 

- For $\Delta < 0$ (negative): The equation has no real roots—the roots are complex numbers.  Since I have not yet studied complex numbers, I have not been able to include the calculation of complex roots; therefore, the program identifies this situation and rejects the input.

- For $\Delta = 0$ (Zero): The equation has two identical real roots (a single real root repeated).

- For $\Delta > 0$ (Positive): The equation has two distinct real roots.

## The Quadratic Formula (Bhaskara's Formula)

When the discriminant confirms the existence of real roots, whether they are identical or distinct ($Δ ≥ 0$), they are calculated using Bhaskara's formula:

$$x = \frac{-b \pm \sqrt{\Delta}}{2a}$$

or:

$$x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}$$


The ± symbol in the formula is implemented computationally as two separate operations, producing the solutions $x_1$ (using +) and $x_2$ (using -).

## Conclusion

By performing these two operations, we obtain the roots of the quadratic equation in question.

# How to use:
- Simply insert the linear coefficient, angular coefficient, and the constant term, according to the general form of the quadratic equation: $ax² + bx + c = 0.$

Quadratic equations with distinct, double, and non-real roots, respectively:

```python
print("Importing module...")
import math
import time
print("Completed")
print()
def coefficient_a():
    a = float(input("Specify the quadratic coefficient (a): "))
    if a == 0:
        print()
        print("\033[31mInvalid structure for quadratic equation (a≠0)\033[0m")
        return coefficient_a()
    else:
        b = float(input("Specify the linear coefficient (b): "))
        c = float(input("Specify the independent term (c): "))
        quadratic_equation = f"{str(a).replace('.', ',')}x² {str(b).replace('.', ',')}x {str(c).replace('.', ',')}"
        if b>0 and c>0:
            quadratic_equation = f"{str(a).replace('.', ',')}x² + {str(b).replace('.', ',')}x + {str(c).replace('.', ',')}"
        if b>0 and c<0:
            quadratic_equation = f"{str(a).replace('.', ',')}x² + {str(b).replace('.', ',')}x {str(c).replace('.', ',')}"
        if b<0 and c>0:
            quadratic_equation = f"{str(a).replace('.', ',')}x² {str(b).replace('.', ',')}x + {str(c).replace('.', ',')}"
        if b<0 and c<0:
            quadratic_equation = f"{str(a).replace('.', ',')}x² {str(b).replace('.', ',')}x {str(c).replace('.', ',')}"
        if b>0 and c == 0:
            quadratic_equation = f"{str(a).replace('.', ',')}x² + {str(b).replace('.', ',')}x"
        if b<0 and c == 0:
            quadratic_equation = f"{str(a).replace('.', ',')}x² {str(b).replace('.', ',')}x"
        if b == 0 and c == 0:
            quadratic_equation = f"{str(a).replace('.', ',')}x²"
        if b == 0 and c>0:
            quadratic_equation = f"{str(a).replace('.', ',')}x² + {str(c).replace('.', ',')}"
        if b == 0 and c<0:
            quadratic_equation = f"{str(a).replace('.', ',')}x² {str(c).replace('.', ',')}"
        print()
        print(f"Quadratic equation to solve: \033[32m{quadratic_equation}\033[0m")
        print()
        print("Calculating the discriminant...")
        discriminant = (b**2) - 4 * a * c
        time.sleep(1)
        print(f"Discriminant calculated successfully: \033[32m{str(discriminant).replace('.', ',')}\033[0m")
        if discriminant < 0:
            time.sleep(1)
            print()
            print("\033[31mx doesn't exist in the set of real numbers.\033[0m")
            return coefficient_a()
        if discriminant == 0:
            time.sleep(1)
            print()
            print("It has coincident real roots.")
            print("\033[32mx exists in the set of real numbers.\033[0m")
            print()
            print("Calculating the roots...")
            time.sleep(1)
            root1 = (b * -1 + math.sqrt(discriminant))/(2*a)
            root2 = (b * -1 - math.sqrt(discriminant))/(2*a)
            print("\033[32mRoots calculated successfully.\033[0m")
            print()
            print(f"Root 1 (+): \033[32m{str(root1).replace('.', ',')}\033[0m")
            print(f"Root 2 (-): \033[32m{str(root2).replace('.', ',')}\033[0m")
            print("Enter the next equation:")
            return coefficient_a()
        if discriminant > 0:
            time.sleep(1)
            print()
            print("\033[32mx exists in the set of real numbers.\033[0m")
            print()
            print("Calculating the roots...")
            time.sleep(1)
            root1 = (b * -1 + math.sqrt(discriminant))/(2*a)
            root2 = (b * -1 - math.sqrt(discriminant))/(2*a)
            print("\033[32mRoots calculated successfully.\033[0m")
            print()
            print(f"Root 1 (+): \033[32m{str(root1).replace('.', ',')}\033[0m")
            print(f"Root 2 (-): \033[32m{str(root2).replace('.', ',')}\033[0m")
            print("Enter the next equation:")
            return coefficient_a()
coefficient_a()
```

# Interpretation

It can be observed that for a negative discriminant, the roots are non-real (complex), and for a positive or zero discriminant, the roots are real. Correspondingly, for $\Delta > 0$ and $\Delta = 0$, the roots are distinct and coincident, respectively.
