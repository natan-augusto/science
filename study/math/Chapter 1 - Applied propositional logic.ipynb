## Theory
### 2.1 Proposition 
A proposition is, by definition, any declarative sentence that can be classified as **true** or **false**.  
A proposition must:
- Be a statement.
- Be declarative.
- Have exactly one logical value (never both at the same time):
    - True (T)
    - False (F)

Example:  
$$p: 7 > 3 \rightarrow V$$
$$p: \sqrt{-1} \quad \exists \quad \mathbb{R} \rightarrow F$$

### 2.2 Negations
The negation is nothing more than the statement that contradicts the main proposition. For every proposition $p$, there is a corresponding negation $\neg p$ (upon further research, I found a notation more appropriate than the “$~p$” mentioned in the book) called “the negation of p”.  
Example:
$$p: 7 > 3$$
$$\neg p: 7 \le 3$$

You can see this relationship in the truth table:
| $p$ | $\neg p$ |
| :--- | ---: |
| $V$ | $F$ |
| $F$ | $V$ |
---

```python
import pandas
proposition = input(print("Enter a mathematical proposition (p): "))
proposition_value = input(print("Enter the truth value of the proposition (T/F): "))
print()
if proposition_value == "T":
    print(f'The proposition "{proposition}" is true.')
    truth_table = {
    "p": [proposition_value],
    "¬p": ["F"],
    }
else:
    print(f"The proposition '{proposition}' is false.")
    truth_table = {
    "p": [proposition_value],
    "¬p": ["T"],
    }
print()
print("Truth table of the proposition: ")
df = pandas.DataFrame(truth_table)
print()
print(df)
print()
print(f"p: {proposition}.")
print(f'¬p: "{proposition}" is false.')
```

# Interpretation
It can be seen that every proposition $p$ has a corresponding negation of $p$ ($\neg p$), and this negation will always have a value opposite to that of the proposition $p$.

# Conclusion
If $p$ is true, then $\neg p$ is false; and if $p$ is false, then $\neg p$ is true in all cases.
